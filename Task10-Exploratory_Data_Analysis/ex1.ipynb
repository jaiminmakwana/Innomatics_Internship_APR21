### AMEO_2015_Data

1. The dataset was released by Aspiring Minds from the Aspiring Mind Employment Outcome 2015 (AMEO).
2. The study is primarily limited  only to students with engineering disciplines. 
3. The dataset contains the employment outcomes of engineering graduates as dependent variables (Salary, Job Titles, and Job Locations) along with the standardized scores from three different areas – cognitive skills, technical skills and personality skills.
4. The dataset also contains demographic features.
5. The dataset  contains  around  40 independent variables and 4000 data points. 
6. The independent variables are both continuous and categorical in nature.
7. The dataset contains a unique identifier for each candidate.

### Is there a relationship between gender and specialisation? (i.e. Does the preference of Specialisation depend on the Gender?)

### Chi square test

Lets make a bold Claim that jobs as a Programming Analyst, Software Engineer,
Hardware Engineer and Associate Engineer  can earn up to 2.5-3 lakhs as a fresh
graduate.

Step - 1:
Alternate Hypothesis: $ H_1:  $ There is a relationship between Gender and Specialization

Null Hypothesis:$ H_0: $ There is no relationship between Gender and Specialization 

Step - 2:

Collect the sample of size n
Compute the sample frequencies

Step - 3: Compute $ \chi^2 $  test statistic

Now you need to check, if the difference in the observed and expected frequencies is too extreme to reject the NULL hypothesis.

Have a look at Observed Frequencies (in the sample)

Compute the Expected Frequencies (under null hyp assumption)<center>$ Expected  Value = \frac{row total * col  total}{grand total}$</center>

Now, test statistic can be computed using below mentioned formula:<center> $ \chi^2 = \Sigma \frac{(o-e)^2}{e} $ </center>


Step - 4: Decide $ \alpha $ and $ df = (rows-1)(cols-1) $

Step - 5: Apply decision rule

Chi Square Test  if <center> $ \chi^2 \gt \chi^2_{df, \alpha} \Rightarrow Reject H_0 $ </center>

p-value Test <center> $ p  value = (1.0 - cdf(test statistic)) $</center>

 Now,<center>  $ if (p  value \lt \alpha) \Rightarrow Accept H_1  or  Reject H_0 $ </center>


```python
data = pd.read_csv('aspiring_minds_employability_outcomes_2015.csv')
pd.crosstab(  data.Specialization,data.Gender, margins=True)
```

```python
observed = pd.crosstab(data.Specialization,data.Gender)

observed
```

```python
from scipy.stats import chi2
from scipy.stats import chi2_contingency
chi2_contingency(observed)
```

```python
# Computing chi2 test statistic, p-value, degree of freedoms

chi2_test_stat = chi2_contingency(observed)[0]
pval = chi2_contingency(observed)[1]
df = chi2_contingency(observed)[2]
```

```python
confidence_level = 0.90

alpha = 1 - confidence_level

chi2_critical = chi2.ppf(1 - alpha, df)

chi2_critical
```

```python
# Ploting the chi2 distribution to visualise

# Defining the x minimum and x maximum
x_min = 0
x_max = 100

# Ploting the graph and setting the x limits
x = np.linspace(x_min, x_max, 100)
y = chi2.pdf(x, df)
plt.xlim(x_min, x_max)
plt.plot(x, y)


# Setting Chi2 Critical value 
chi2_critical_right = chi2_critical

# Shading the right rejection region
x1 = np.linspace(chi2_critical_right, x_max, 100)
y1 = chi2.pdf(x1, df)
plt.fill_between(x1, y1, color='orange')
```

```python
if(chi2_test_stat > chi2_critical):
    print("Reject Null Hypothesis")
else:
    print("Fail to Reject Null Hypothesis")
```

```python
if(pval < alpha):
    print("Reject Null Hypothesis")
else:
    print("Fail to Reject Null Hypothesis")
```

### Conclusion: 
Thus from $\chi^2$ test there is a relationship between Gender and Specialization

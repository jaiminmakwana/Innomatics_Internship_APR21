```python
# here we can see that -1 appears for 461 times, which is a big number. I will replace it with most frequent value.

df.JobCity.replace(-1,df.JobCity.mode()[0],inplace=True)
df.head()
```

```python
# here we can see that -1 appears for 246 times, which is a big number. I will replace it with most frequent value.

df.Domain.replace(-1,df.Domain.mean(),inplace=True)
df.head()
```

```python
# here we can see that -1 appears for 246 times, which is a big number. I will replace it with most frequent value.

df.ComputerProgramming.replace(-1, df.ComputerProgramming.mean(), inplace=True)
df.ElectronicsAndSemicon.replace(-1, df.ElectronicsAndSemicon.mean(), inplace=True)
df.ComputerScience.replace(-1, df.ComputerScience.mean(), inplace=True)
df.MechanicalEngg.replace(-1, df.MechanicalEngg.mean(), inplace=True)
df.ElectricalEngg.replace(-1, df.ElectricalEngg.mean(), inplace=True)
df.TelecomEngg.replace(-1, df.TelecomEngg.mean(), inplace=True)
df.CivilEngg.replace(-1, df.CivilEngg.mean(), inplace=True)
df.head()
```

```python
typ =['int64', 'float64']
num=[]
cat =[]
for i in df.columns:
    if df[i].dtype.name in typ:
        num.append(i)
    if df[i].dtype.name == 'object':
        cat.append(i)
        
print(num)
print(cat)
```

```python
for i in num:
    sns.set_style('whitegrid')
    sns.histplot(x=df[i], bins = 30, legend=True)    
    plt.show()

for i in num:
    sns.kdeplot(x=i, data=df, legend=True)
    plt.show() 
```
